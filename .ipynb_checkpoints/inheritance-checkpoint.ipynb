## INHERITANCE using LOGIG GATES EXAMPLE
Consider the following hierarchial structure:-


We need the class `LogicGate` to do the following:-
- identify itself
- allow the user to select a gate
- provide output

```python
class LogicGate:
    def __init__(self, gate):
        self.label = gate
        self.output = None
     
    def getLabel(self):
        return self.label
    
    def getOutput(self):
        self.output = self.gateLogic()
        return self.output
```

If you notice the figure, it's evident we can classify binary and unary gate on the basis of the no. of input pins it has.

---
The constructors in both of these classes start with an explicit call to the constructor of the parent class using the parentâ€™s `__init__` method.

When creating an instance of the `BinaryGate` class, we first want to initialize any data items that are inherited from `LogicGate`. In this case, that means the label for the gate.

The only behavior that the `BinaryGate` class adds is the ability to get the values from the two input lines. Since these values come from some external place, we will simply ask the user via an input statement to provide them. The same implementation occurs for the `UnaryGate` class except that there is only one input line.

---
**super**:
Python also has a function called `super` which can be used in place of explicitly naming the parent class. This is a more general mechanism, and is widely used, especially when a class has more than one parent. 


```python
class BinaryGate(LogicGate):
    def __init__(self, gate):
        LogicGate.__init__(self, gate)
        
        self.pinA = None
        self.pinB = None
        
    def getPinA(self):
        return (int(input("Enter value for Pin A: " + self.getLabel() + "-->")))
    
    def getPinB(self):
        return (int(input("Enter value for Pin B: " + self.getLabel() + "-->")))
    

class UnaryGate(LogicGate):
    def __init__(self, gate):
        LogicGate.__init__(self, gate)
        
        self.pinA = None
        self.pinB = None
        
    def getPin(self):
        return (int(input("Enter value for Pin: " + self.getLabel() + "-->")))
    
```
