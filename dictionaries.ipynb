```python
## dictionary to data frame
## Data Frame is an imp. data str. A way to store tabular data where you can label the rows and the columns. 
## One way to build a DataFrame is from a dictionary.

## remember textual to tabular data

#import pandas
import pandas as pd

## define the keys and values

##** NOTE: the values are an array, so beware the braces!
## key = country and value = country_names
country_names = ['USA','UK','China','India','France'];

## key = driving_side_right and value = driving_side_value is a boolean list
driving_side_value = [True, False, True, False, True];

## key = carsper1000people and value = cpt_value
cpt_value = [900,800,600,500,750];

## custom dictionary
my_cars_dict = {'country':country_names, 
                'driving_side_right':driving_side_value, 
                'carsper1000people':cpt_value};

## **BUILD DATA FRAME
cars = pd.DataFrame(my_cars_dict)
print(cars);
```

```python
## previously the row labels went from 0 to 4. these can be renamed, apparently, by creating a list
row_labels = ['US','UK','CHN','IND','FRA'];

import pandas as pd
## key = country and value = country_names
country_names = ['USA','UK','China','India','France'];

## key = driving_side_right and value = driving_side_value is a boolean list
driving_side_value = [True, False, True, False, True];

## key = carsper1000people and value = cpt_value
cpt_value = [900,800,600,500,750];

## custom dictionary
my_cars_dict = {'country':country_names, 
                'driving_side_right':driving_side_value, 
                'carsper1000people':cpt_value};

## **BUILD DATA FRAME
cars = pd.DataFrame(my_cars_dict)
print(cars);

cars.index = row_labels;
print(cars);
```

```python
## using loc and iloc
## loc is label based operation to find anything from a data frame, while iloc is integer-index based.
## => you gotta specify the rows and columns.

import pandas as pd
row_labels = ['US','UK','CHN','IND','FRA'];

import pandas as pd
## key = country and value = country_names
country_names = ['USA','UK','China','India','France'];

## key = driving_side_right and value = driving_side_value is a boolean list
driving_side_value = [True, False, True, False, True];

## key = carsper1000people and value = cpt_value
cpt_value = [900,800,600,500,750];

## custom dictionary
my_cars_dict = {'country':country_names, 
                'driving_side_right':driving_side_value, 
                'carsper1000people':cpt_value};

## **BUILD DATA FRAME
cars = pd.DataFrame(my_cars_dict)
cars.index = row_labels;

print(cars);
print('\n')
print(cars.loc['UK']);
print('\n')
print(cars.iloc[3]);
print('\n')
print(cars.loc[['IND','CHN']])
print('\n');
print(cars.iloc[[0,1,3]])
print('\n')
print(cars.loc[:,['country']])
print('\n')
print(cars.loc[:,['country','driving_side_right']])

```
